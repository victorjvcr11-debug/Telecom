#**Cargar .JSON**

```python
#Importar biblioteca Pandas
import pandas as pd
#Importando el modulo JSON
import json
```

```python
data ='/content/drive/MyDrive/Colab Notebooks/TelecomX_Data.json'
#Leyendo el archivo JSON usando el modulo Python JSON
with open(data,'r') as f:
  datos = json.loads(f.read())
datos
```

```python
#Normalizar los datos con los parametros con record_path y meta
datos= pd.json_normalize(datos)
```

```python

datos
```

```python
#Poder ver todas las columnas completas
pd.set_option('display.max_columns',None)
```

```python
datos
```

```python
#Quitar todo antes del .
datos.columns = datos.columns.str.replace('^.*?\.', '',regex=True)
```

```python
datos
```

```python
#Quitar todo antesd del .
datos.columns = datos.columns.str.replace('^.*?\.', '',regex=True)
```

```python
datos
```

```python
#Poner la inicial en minisculas
datos.columns = datos.columns.str.replace(r'(?<=[a-z])(?=[A-Z])', '_', regex=True)
datos.columns = datos.columns.str.lower()
```

##Tratar Floats

```python
import numpy as np
```

```python
#Quitar espacios ya que asi no permite hacer la transfomacion
datos['total'] = datos['total'].replace(' ','').replace('',np.nan)
#Cambiar Object por Float
datos['total'] = datos['total'].astype(np.float64)
```

```python
datos.info()
```

```python
datos['cuentas_diarias'] = datos['monthly'].div(30)

```
